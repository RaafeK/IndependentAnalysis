```python
X = df2[['Volume','Temp','Prices','Month']]
Y = df2['Adj_Close']
```

```python
from sklearn import tree
from sklearn import neural_network
from sklearn.model_selection import cross_val_score
```

```python
reg = tree.DecisionTreeRegressor(min_samples_leaf=2)
reg.fit(X,Y)
```

```python
reg.score(X,Y)
```

```python
reg.predict(X)
```

```python
plt.plot(reg.predict(X),'--r')
plt.legend('Predicted')
plt.plot(df2['Close'].values,'-b')
plt.plt
```

```python
np.average(cross_val_score(reg,X,Y,cv=10))
```
